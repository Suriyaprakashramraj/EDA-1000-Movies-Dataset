# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
data['Revenue (Millions)'].value_counts(ascending = True)
```

```python
data['Revenue (Millions)'].unique()                                   #Finding the unique values in the Revenue variable
```

```python
data.rename(columns = {'Runtime (Minutes)': 'Runtime', 'Revenue (Millions)': 'Revenue'}, inplace = True)    
```

```python
data[~(data['Revenue'] == '-')]['Revenue'].median()
```

```python
data['Revenue'].fillna(float('47.985'), inplace = True)              #Replacing the NaN values in the Revenue variable with the Median Value
```

```python
data['Revenue'].unique()                                              
```

```python
data[~(data['Metascore'] == '-')]['Metascore'].median()           #Replacing the NaN values in the Metascore variable with the Median Value          
```

```python
data['Metascore'].fillna(float('59.5'), inplace = True)
```

```python
data['Metascore'].unique()                                 #Finding the unique values in the Metascore variable after filling Nan values                                           
```

```python
print(data['Rating'].sum())
print(data['Votes'].sum())
```

```python
Rating_Score_Weight = data['Rating'].sum()/data['Votes'].sum()                    #Finding the Wighted Average for the 'Rating' and 'Votes' variables
Rating_Score_Weight
```

```python
data['Rating*Votes'] = data['Rating']*data['Votes']             #creating a New Feature by multiplying 'Rating' and 'Votes' variables
```

```python
data['Ranking_Factor'] = data['Rating*Votes']*Rating_Score_Weight  #creating a New Feature by multiplying ''Rating*Votes'' and the Wighted Average variables
```

```python
data = data[ ["Rank", "Title", "Genre", "Description", "Director", "Actors", "Year", "Runtime", "Rating", "Votes", "Rating*Votes", "Ranking_Factor", "Revenue", "Metascore"]]
print(data.shape)
data.head(2)
```

```python
data.info()
```

```python
#creating a new dataframe (DataFrame 2) with required varibales only
data_new = data[['Title', 'Genre', 'Director', 'Actors', 'Year', 'Runtime', 'Revenue', 'Rating', 'Votes', 'Ranking_Factor', 'Metascore']]
data_new.head(1)
```

```python
#creating a new dataframe (DataFrame 3) for splitting Genres with multiple genre items
genre_split = data_new['Genre'].str.split(',').apply(pd.Series, 1).stack()                 #Splitting and stacking the multiple genre items
genre_split.index = genre_split.index.droplevel(-1)
genre_split.name = 'Genre'
del data_new['Genre']
data_split_genre = data_new.join(genre_split)
data_split_genre                                                                           # New dataframe with splitted genre items
```

```python
data_split_genre['Genre'].unique()
```

```python
print(data_split_genre.shape)
data_split_genre.head(5)
```

```python
#creating a new dataframe (DataFrame 4) for splitting and stacking the Actors column cells 
actors_split = data_new['Actors'].str.split(',').apply(pd.Series, 1).stack()
actors_split.index = actors_split.index.droplevel(-1)
actors_split.name = 'Actors_Split'
del data_new['Actors']
data_split_actors = data_new.join(actors_split)
data_split_actors                                                            # New dataframe with splitted Actors items
```

**8.4.2 Which interval of Runtime generated higher revenues and are critically acclaimed?**

```python
data['Runtime'].describe()
```

Movies' Runtime will be categorised into different levels based on the time duration as follows:
> | Runtime(in minutes) | Level     |
|---------|-----------------|
|   66-100  | Short|
| 100-111   |Medium|
| 111-123   | Moderately Long|
| 123-191   | Long|

```python
runtime_bin_edges = [66, 100, 111, 123, 191]                                
runtime_bin_names = ['Short', 'Medium', 'Moderately Long', 'Long']                    #creating bins on the 5-point summary for the runtime
data_new['Runtime_levels'] = pd.cut(data_new.Runtime, runtime_bin_edges, labels=runtime_bin_names)
data_new.head()
```

```python
 #creating a dataframe comparing the runtime levels against Revenue, Rating, and Metascore
runtime_stats = data_new.groupby('Runtime_levels')['Revenue', 'Rating', 'Metascore'].mean() 
runtime_stats
```

```python
ax1, ax2, ax3 = runtime_stats.plot.bar(subplots=True);
fig = ax1.get_figure()
fig.set_size_inches(10, 10)
ax1.set_title("")
ax2.set_title("")
ax3.set_title("")
ax3.set_xticklabels(runtime_stats.index, rotation = 45);
```

- Undoubtedly, Movies with **longest runtime** genrated higher revenue when comparaed to Short, Medium, and Moderately Long movies. <b> 
- On the other hand, though Movies with longest runtime were critically acclaimed slightly higher than other categories, there is no intersting trend in terms of runtime vs movie rating.
